### h.py — Lotto 6개 중복없이 맞추기
**문제**
- 6개의 random number를 중복 없이 생성
- 6개를 모두 맞출 때까지 입력을 받는다

````python
import random

L=[]
while len(L) < 6:
    n = random.randint(0,9)
    if n not in L:
        L.append(n)

print('Lotto numbers:', L)

user = []
while len(user) < 6:
    u= int(input("로또 6개를 맞춰주세요."))
    if u in L and u not in user:
        print('로또 %d를 맞췄습니다.' % u)
        user.append(u)
        print("현재까지 맞춘 로또 : ", user)
    else :
        print('%d는 로또 수에 없습니다.' %u)

print("모두 맞췄습니다.")
````

### g.py — Password 조합 확인
 
**문제**
- 문자열 하나를 입력받아 영문자만, 숫자만, 영문자+숫자만으로 구성됐는지 확인
- 세 조건 모두 아니면 특수기호 포함 → 유효한 패스워드

````python
password = input('Password: ')

has_alpha = False
has_digit = False
has_special = False

for c in password:
    if c.isalpha():
        has_alpha == True
    if c.isdigit():
        has_digit = True
    if not c.isalnum():
        has_special = True

if not has_alpha and not has_special:
    print('숫자로만 작성됨 - 유효하지 않음')
elif not has_digit and not has_special:
    print('영문자로만 작성됨 - 유효하지 않음')
elif not has_special:
    print('영문자+숫자로만 작성됨 - 유효하지 않음')
else:
    print('영문자, 숫자, 특수기호 포함 - 유효한 패스워드')
````

### n.py — 홀수만 모으기 (앞/뒤)
 
**문제**
- 0~99 사이 숫자 50개 리스트에서 홀수만 모으기
- (1) 앞에서부터, (2) 뒤에서부터

````python
forward_L = []
reverse_L = []

for i in range(100):
    if i % 2 == 1 :
        forward_L.append(i)
    if (99-i) % 2 == 1:
        reverse_L.append(99-i)

print("forward_L : ", forward_L)
print("reverse_L : ", reverse_L)
````

### p.py — 두 리스트 교집합 / 합집합 (중복 없이)
 
**문제**
- 0~99 사이 50개 리스트 2개 생성
- (1) 공통 원소 리스트
- (2) 중복 피해 합치기

````python
import random

A = []
B = []

for i in range(50):
    a = random.randint(0,99)
    b = random.randint(0,99)
    A.append(a)
    B.append(b)

print("A : ", A)
print("B : ", B)
common = []
for v in A : 
    if v in A and v in B and v not in common:
        common.append(v)
print("common : ", common)

union = A
for v in B:
    if v not in A and v in B:
        union.append(v)

print("union : ", union)
````

### q.py — 홀수·짝수 번갈아 나열
 
**문제**
- 0~99 사이 10개 리스트 → 정렬 후 홀수, 짝수, 홀수, 짝수 순으로 재배열
- 예: `[1, 11, 14, 26, 44, 51, 63, 78, 82, 85]` → `[1, 14, 11, 26, 51, 44, 63, 78, 85, 82]`

````python
import random

x = []
for i in range(10):
    x.append(random.randint(0, 99))

Y = sorted(x)
print('정렬:', Y)

odds=[]
evens = []

for v in Y:
    if v % 2 == 1:
        odds.append(v)
    else : 
        evens.append(v)
print("홀수 : ", odds)
print("짝수 : ", evens)

result = []
i = 0
j = 0
while i < len(odds) and j < len(evens):
    result.append(odds[i])
    result.append(evens[j])
    i += 1
    j += 1

# 남은 것 추가
while i < len(odds):
    result.append(odds[i])
    i += 1
while j < len(evens):
    result.append(evens[j])
    j += 1

print('번갈아:', result)
````

#### i1.py — factorial 함수
 
**문제**: n의 factorial을 계산하는 함수

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input('number: '))
print('factorial of', n, 'is', factorial(n))
````

#### i2.py — 소수 판별 함수
 
**문제**: n이 소수인지 확인하는 함수 (소수: 1과 자신만으로 나눠지는 수)

````python
def is_prime(n):
    if n < 2:
        return 0
    for i in range(2, n):
        if n % i == 0:
            return 0
    return 1

n = int(input('number: '))
print(is_prime(n))
````

#### i3.py — 문자열 리스트 → 각 원소 글자 수 반환
 
**문제**: string 리스트를 받아 각 원소별 글자 수 리스트를 return

````python
def count_letters(words):
    result = []
    for w in words:
        result.append(len(w))
    return result

words = ['cat', 'window', 'demonstrate']
print(count_letters(words))
# [3, 6, 11]
````
