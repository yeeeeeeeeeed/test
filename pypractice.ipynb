### h.py — Lotto 6개 중복없이 맞추기
**문제**
- 6개의 random number를 중복 없이 생성
- 6개를 모두 맞출 때까지 입력을 받는다

````python
import random

L=[]
while len(L) < 6:
    n = random.randint(0,9)
    if n not in L:
        L.append(n)

print('Lotto numbers:', L)

user = []
while len(user) < 6:
    u= int(input("로또 6개를 맞춰주세요."))
    if u in L and u not in user:
        print('로또 %d를 맞췄습니다.' % u)
        user.append(u)
        print("현재까지 맞춘 로또 : ", user)
    else :
        print('%d는 로또 수에 없습니다.' %u)

print("모두 맞췄습니다.")
````

### g.py — Password 조합 확인
 
**문제**
- 문자열 하나를 입력받아 영문자만, 숫자만, 영문자+숫자만으로 구성됐는지 확인
- 세 조건 모두 아니면 특수기호 포함 → 유효한 패스워드

````python
password = input('Password: ')

has_alpha = False
has_digit = False
has_special = False

for c in password:
    if c.isalpha():
        has_alpha = True
    if c.isdigit():
        has_digit = True
    if not c.isalnum():
        has_special = True

if not has_alpha and not has_special:
    print('숫자로만 작성됨 - 유효하지 않음')
elif not has_digit and not has_special:
    print('영문자로만 작성됨 - 유효하지 않음')
elif not has_special:
    print('영문자+숫자로만 작성됨 - 유효하지 않음')
else:
    print('영문자, 숫자, 특수기호 포함 - 유효한 패스워드')
````

### q.py — 홀수·짝수 번갈아 나열
 
**문제**
- 0~99 사이 10개 리스트 → 정렬 후 홀수, 짝수, 홀수, 짝수 순으로 재배열
- 예: `[1, 11, 14, 26, 44, 51, 63, 78, 82, 85]` → `[1, 14, 11, 26, 51, 44, 63, 78, 85, 82]`

````python
import random

x = [random.randint(0,99) for _ in range(10)]
x.sort()
print('정렬:', x)
evens = []
odds = []

for v in x:
    if v % 2 == 1 :
        odds.append(v)
    else:
        evens.append(v)

print("홀수 : ", odds)
print("짝수 : ", evens)

result = []
for i in range(min(len(odds), len(evens))):
    result.append(odds[i])
    result.append(evens[i])

if i+1  == len(odds):
    for j in range(i, len(evens)):
        result.append(evens[j])
else : 
    for j in range(i, len(odds)):
        result.append(odds[j])

print('번갈아:', result) 
````

````python
def is_prime(n):
    if n < 2 :
        return 0
    else:
        for i in range(2,n):
            if n%i == 0:
                return 0
        
        return 1

user = int(input("숫자를 입력하세요 : "))
if is_prime(user) : print("%d는 소수입니다." %(user))
else : print("%d는 소수가 아닙니다." %(user))
````

#### i5.py — n 이하의 소수 모음 return

````python
def prime(n):
    result = []
    for i in range(2,n+1):
        is_prime = True
        for j in range(2, i):
            if i%j == 0:
                is_prime = False
                break
        if is_prime : 
            result.append(i)
    
    return result

print(prime(19))
````

#### i6.py — random 10개 합·평균·최대·최소 반환

````python
import random

def stats(L):
    total = sum(L)
    avg = total / len(L)
    maximum = L[0]
    minimum = L[0]
    for v in L:
        if v > maximum:
            maximum = v
        if v < minimum:
            minimum = v
    return total, avg, maximum, minimum

nums = [random.randint(0, 99) for _ in range(10)]
print('리스트:', nums)
total, avg, mx, mn = stats(nums)
print('합:', total, '/ 평균:', avg, '/ 최대:', mx, '/ 최소:', mn)
````

````python
import random

def filter_multiples_of_3(L):
    sorted_L = sorted(L)
    result = []
    for v in sorted_L:
        if v % 3 == 0:
            result.append(v)
    return result

nums = [random.randint(0, 99) for _ in range(30)]
print('원본:', nums)
print('3의배수(정렬):', filter_multiples_of_3(nums))
````

#### m5.py — 가위바위보 우선순위 dictionary로 승부 계산


````python
import random

wins = {
    'scissors': 'paper',
    'rock': 'scissors',
    'paper': 'rock'
}

L = ['scissors', 'rock', 'paper']
win_a = 0
win_b = 0
tie = 0

for i in range(10):
    a = random.choice(L)
    b = random.choice(L)

    print('A:', a, '/ B:', b, end=' -> ')
    if a == b:
        print('tie')

    elif wins[a] == b:
        win_a += 1
        print('A wins')

    else:
        win_b += 1
        print("B wins")

print('A 승리:', win_a, '/ B 승리:', win_b, '/ 무승부:', tie)
````

````python
score = {'Lee': [90, 85], 'Kim': [45, 96], 'Park': [53, 89]}
child = {'Lee': 2, 'Kim': 1, 'Park': 3}

# 평균 성적이 가장 높은 사람
avg_score = {name: sum(score[name]) / len(score[name]) for name in score}
best_score = max(avg_score, key=avg_score.get)

# 자녀 수가 가장 많은 사람
best_child = max(child, key=child.get)

print('성적 최고:', best_score, '/ 자녀수 최고:', best_child)
if best_score == best_child:
    print('같은 사람입니다.')
else:
    print('다른 사람입니다.')
````

#### y3.py — 성적 제일 높은 사람이 공부시간도 제일 많은가

````python
score = {'Lee': [90, 85], 'Kim': [45, 96], 'Park': [53, 89]}
child = {'Lee': 2, 'Kim': 1, 'Park': 3}
hours = {'Lee': 1000, 'Kim': 200, 'Park': 300}
area = {'Lee': 'loc A', 'Kim': 'loc B', 'Park': 'loc A'}
house = [('loc A', 50), ('loc B', 100), ('loc C', 85)]

max_score = 0
max_score_name = ''
max_hours = 0
max_hours_name = ''

for name, s in score.items():
    average = sum(s)/2
    if average > max_score:
        max_score = average
        max_score_name = name


for name, c in hours.items():
    if c > max_hours:
        max_hours = c
        max_hours_name = name

if max_hours_name == max_score_name :
    print(" 같은 사람입니다.")
else:
    print(" 다른 사람 입니다.")
````

````python
score = {'Lee': [90, 85], 'Kim': [45, 96], 'Park': [53, 89]}
hours = {'Lee': 1000, 'Kim': 200, 'Park': 300}

avg_score = {name: sum(score[name]) / len(score[name]) for name in score}
best_score = max(avg_score, key=avg_score.get)
best_hours = max(hours, key=hours.get)

print('성적 최고:', best_score, '/', avg_score[best_score])
print('공부시간 최고:', best_hours, '/', hours[best_hours])

if best_score == best_hours:
    print('성적이 높은 사람이 공부시간도 가장 많습니다.')
else:
    print('성적이 높은 사람과 공부시간이 많은 사람이 다릅니다.')
````
