### q1.py — 3의 배수 / 5의 배수 개수 세기
 
**문제**
1~100 범위의 random number 10개 리스트를 만들고, 그 중 3의 배수 개수와 5의 배수 개수를 출력하시오.
 
**힌트 (슬라이드 제공)**
```python
import random
 
# while (반복조건):
#     num = random.randint(1, 100)
```
 
**실행 예시**
```
list : [90, 78, 71, 39, 19, 89, 92, 95, 70, 27]
3의 배수 개수 : 4
5의 배수 개수 : 3
```

````python
import random

L = []
while len(L) < 10:
    num = random.randint(1, 100)
    L.append(num)

print('list :', L)

count3 = 0
count5 = 0
for n in L:
    if n % 3 == 0:
        count3 += 1
    if n % 5 == 0:
        count5 += 1

print('3의 배수 개수 :', count3)
print('5의 배수 개수 :', count5)
````

### q2.py — 합이 20 넘을 때 중지, 중간값 출력
 
**문제**
10보다 작은 정수 N을 계속 입력받아 리스트에 저장한다. 리스트 안의 숫자의 합이 20보다 커지면 중지하고, 리스트의 개수와 중간에 위치한 수를 출력하시오.
- 개수가 11개이면 6번째(인덱스 5), 개수가 10개이면 5번째(인덱스 4) → 개수 // 2 번째

````python
L = []
while True:
    N = int(input('N (10보다 작은 정수): '))
    L.append(N)
    if sum(L) > 20:
        break

count = len(L)
mid = L[count // 2]
print('개수 :', count)
print('중간값 :', mid)
````

### q2.py — 두 리스트 모두에 없는 원소 찾기
 
**문제**
random number(1~30)로 구성된 중복 없는 리스트 A, B를 각각 10개씩 생성한다. 1~30 중 A와 B 둘 다에 속하지 않는 원소들로 리스트 C를 만드시오.

````python
import random

A = []
while len(A) < 10:
    n = random.randint(1, 30)
    if n not in A:
        A.append(n)

B = []
while len(B) < 10:
    n = random.randint(1, 30)
    if n not in B:
        B.append(n)

print('A:', A)
print('B:', B)

C = []
for n in range(1, 31):
    if n not in A and n not in B:
        C.append(n)

print('C:', C)
````

### q1.py — 약수의 합 함수
 
**문제**
임의의 숫자 n이 주어지면 자기자신 포함 모든 약수의 합을 출력하는 프로그램을 작성하시오.
- 약수를 구하고 합을 return하는 함수 `divisor(n)` 작성
- 출력은 함수 밖에서 할 것
**실행 예시**
```
6의 약수: (1, 2, 3, 6) → 합: 12
```

````python
def divisor(n):
    total = 0
    for i in range(1, n + 1):
        if n % i == 0:
            total += i
    return total

n = int(input('n: '))
result = divisor(n)
print('약수의 합:', result)
````

### q2.py — Euclidean distance 함수
 
**문제**
좌표 2개 (x1, y1), (x2, y2)를 입력받아 Euclidean distance를 구하시오.
- 함수 `euclidian(x1, y1, x2, y2)` 작성, 계산 결과를 return
- 거리는 소수 둘째자리까지 출력
- 출력은 함수 밖에서 할 것
**공식**: `((x1 - x2)**2 + (y1 - y2)**2) ** 0.5`

````python
def euclidian(x1, y1, x2, y2):
    return ((x1 - x2)**2 + (y1 - y2)**2) ** 0.5

x1 = float(input('x1: '))
y1 = float(input('y1: '))
x2 = float(input('x2: '))
y2 = float(input('y2: '))

result = euclidian(x1, y1, x2, y2)
print('거리: %.2f' % result)
````

### q3.py — 홀수/짝수 판별 문자열 리스트 함수
 
**문제**
random number(1~99) 10개 리스트 L을 만들고, 각 숫자가 홀수인지 짝수인지 판별해 "odd"/"even" 문자열 리스트를 출력하시오.
- 함수 `makelist(L)`: L을 받아 "odd"/"even" 리스트 A를 return
- 출력은 함수 밖에서 할 것
**실행 예시**
```
L = [16, 59, 55, 39, 36, 33, 96, 20, 93, 73]
A = ['even', 'odd', 'odd', 'odd', 'even', 'odd', 'even', 'even', 'odd', 'odd']
```

````python
import random

def makelist(L):
    A = []
    for n in L:
        if n % 2 == 1:
            A.append('odd')
        else:
            A.append('even')
    return A

L = []
for i in range(10):
    L.append(random.randint(1, 99))

A = makelist(L)
print('L =', L)
print('A =', A)
````

### q1.py — 중복 없는 5개 random, max 찾기 (내장함수 금지)
 
**문제**
1~100 사이 서로 다른 정수 5개를 생성한 후 가장 큰 숫자를 출력하시오.
- `max()`, `sort()` 등 내장함수 사용 금지
- 함수 1: 중복 없는 random 숫자 5개 생성 후 리스트 return
- 함수 2: 숫자 두 개를 받아 더 큰 값 return

````python
import random

def make_random_list():
    L = []
    while len(L) < 5:
        n = random.randint(1, 100)
        if n not in L:
            L.append(n)
    return L

def bigger(a, b):
    if a >= b:
        return a
    else:
        return b

L = make_random_list()
print('리스트:', L)

maximum = L[0]
for i in range(1, len(L)):
    maximum = bigger(maximum, L[i])

print('최대값:', maximum)
````

### q2.py — 모음/자음 출력 함수
 
**문제**
문자열을 입력받아, 모음(a, e, i, o, u)은 그대로, 자음은 '-'로 출력하시오.
- 함수: 문자열을 순서대로 확인해 모음이면 출력하고 아니면 '-' 출력, 모음의 개수를 return
**실행 예시**
```
Input string : apple
a - - - e
Count : 2
```

````python
def check_vowel(s):
    vowels = 'aeiou'
    count = 0
    for c in s:
        if c in vowels:
            print(c, end=' ')
            count += 1
        else:
            print('-', end=' ')
    print()
    return count

s = input('Input string : ')
cnt = check_vowel(s)
print('Count :', cnt)
````
