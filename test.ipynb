**출력 예시**
```
알파벳 입력 : t

---
알파벳 입력 : z

```

````python
ch = input("알파벳 입력 :")
char = 'python'
if ch in char : 
    print("포함되어 있습니다")

    for i in range(len(char)):
        if ch == char[i]:
            num = i+1
            print(f"'{ch}'는 '{char}'의 {num}번째 글자입니다.")
else:
    print("포함되어 있지 않습니다")   
````

### 문제 7
 
> `while`문만 사용하여 1부터 N까지의 합과 곱을 구하시오.
> N은 10 이하의 자연수를 입력받는다.
 
**조건**
- `while` 문만 사용 (for 사용 불가)
- 합(sum)과 곱(product) 모두 출력
**출력 예시**
```
Input N : 5
sum     = 15
product = 120
```

````python
n = int(input("Input N : "))
i = 1
sum = 0
product = 1
while True :
    if i < n+1 :
        sum+=i
        product *= i
        i+=1
    else:
        break

print("sum : ", sum)
print("product : ", product)
````

### 문제 9
 
> `while`과 `break`를 이용하여 N 이하의 자연수 중 가장 큰 소수를 구하시오.
> (N은 50 이하 입력)
 
**조건**
- `while` 문과 `break` 사용 필수
- N 이하에서 가장 큰 소수 출력
**출력 예시**
```
Input N : 20
20 이하 가장 큰 소수 : 19
```

````python
n = int(input("Input N : "))
result=[]
print("Input N : ", n)
if n < 2 :
    print("소수가 아닙니다.")
else :
    can = 2
    while can <= n :
        is_prime  = True
        i=2
        while i < can:
            if can % i == 0 :
                is_prime = False
                break
            i+=1
        if is_prime : 
            result.append(can)
        
        can+=1
    

    print(f'{n} 이하 소수 : {result}')

````

> N을 입력받아 N번째 줄까지 아래처럼 `*` 직각삼각형을 출력하시오.

### 문제 11
 
> 숫자 맞추기 게임을 작성하시오.
> `random`으로 1~100 사이의 숫자를 뽑고, 사용자가 맞출 때까지 입력받는다.
> 입력값이 정답보다 크면 `Too high`, 작으면 `Too low`, 맞으면 `Correct!`와 시도 횟수 출력.
 
**조건**
- `random.randint(1, 100)` 사용
- `while True` + `break` 사용
- 시도 횟수 카운트
**출력 예시**
```
Input your guess : 50
Too high
Input your guess : 25
Too low
Input your guess : 37
Correct!  시도 횟수 : 3
```

````python
import random
n = random.randint(1,100)
print(n)
cnt = 0
while True :
    g = int(input("Input your guess : "))
    print("Input your guess : ",g)
    cnt+=1
    if g < n :
        print("Too low")
    elif g > n :
        print("Too high")
    else:
        print("Correct!")
        break

print("시도 횟수 :", cnt)


````

> `random(1~100)`으로 중복 없이 숫자 8개를 뽑아 리스트를 만들고
> 최솟값, 최댓값, 합계, 평균을 출력하시오.

- 중복 없는 랜덤 리스트 생성 (`if num not in list` 패턴)
- 평균은 소수점 2자리로 출력
**출력 예시**
```
list : [45, 12, 88, 33, 67, 5, 71, 29]
최솟값 : 5
최댓값 : 88
합계   : 350
평균   : 43.75
```

````python
import random
del sum
random_list = []

while len(random_list) < 8:
    n  = random.randint(1,100)
    if n  not in random_list:
        random_list.append(n)

print(random_list)
print("최솟값",  min(random_list))
print("최대값",  max(random_list))
print('합계   :', sum(random_list))
print(f'평균   : {sum(random_list)/len(random_list):.2f}')
````

### 문제 13
 
> `random(1~50)`으로 중복 없이 10개를 뽑아 리스트를 만들고
> 오름차순 정렬한 리스트와 내림차순 정렬한 리스트를 출력하시오.
 
**조건**
- 중복 없는 랜덤 리스트 생성
- `.sort()` 또는 `sorted()` 사용
- 오름차순 / 내림차순 둘 다 출력
**출력 예시**
```
original : [34, 7, 21, 48, 3, 19, 42, 11, 27, 5]
오름차순  : [3, 5, 7, 11, 19, 21, 27, 34, 42, 48]
내림차순  : [48, 42, 34, 27, 21, 19, 11, 7, 5, 3]
```

````python
import random
result = []

while len(result) < 10:        # for 대신 while로 중복 제거
    a = random.randint(1, 50)
    if a not in result:
        result.append(a)

a = sorted(result)             # 오름차순 새 리스트 반환
b = sorted(result, reverse=True)  # 내림차순 새 리스트 반환

print('original :', result)
print('오름차순  :', a)
print('내림차순  :', b)
````
